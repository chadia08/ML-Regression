**Import Data**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv('dataset/CarPrice_Assignment.csv')
data.head()
```

```python
# Checking Outlier

sns.boxplot(data['price'])
plt.title("Outliers Variable Distribution", fontsize = 14, fontweight = 'bold')
plt.ylabel("Price Range", fontweight = 'bold')
plt.xlabel("Continuous Variable", fontweight = 'bold')

#it seems that there is a lot of outliers
```

```python
#drop insignificant features
data.drop('car_ID',axis=1)
```

```python
# Extracts the car brand from the 'CarName' column by taking the first part of each name.
data['CarName'] = data['CarName'].str.split(' ',expand=True)[0]
```

```python
# Handling duplicate values in car name
data['CarName'].unique()
```

**<ul>Handling duplicate values in car name**
    <li>nissan and Nissan are same</li>
    <li>toyota and toyouta are same</li>
    <li>vokswagen , volkswagen and vw are same</li>
    <li>mazda and maxda are same</li>
    <li>porcshce and porsche are same</li></ul>



```python
data['CarName'] = data['CarName'].replace({'maxda': 'mazda', 'nissan': 'Nissan', 'porcshce': 'porsche', 'toyouta': 'toyota', 
'vokswagen': 'volkswagen', 'vw': 'volkswagen'})
```

```python
data['symboling']=data['symboling'].astype('str')
```

```python
# Segregation of Numerical and Categorical Variables/Columns
cat_col = data.select_dtypes(include=['object']).columns
num_col = data.select_dtypes(exclude=['object']).columns
```
